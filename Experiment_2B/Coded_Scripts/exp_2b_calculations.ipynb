```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Performing the required calculations from the following cells.

```python
# Applying linear regression in TLS sense
def linear_regression(x, y):
    covxy = np.cov(x, y, bias=True)
    xmean = np.mean(x)
    ymean = np.mean(y)

    # using TLS solution
    Sxx, Sxy, Syy = covxy[0][0], covxy[1][0], covxy[1][1]
    alpha = ((Syy - Sxx) + np.sqrt((Syy - Sxx)**2 + 4*Sxy**2))/(2*Sxy)
    beta = ymean - alpha*xmean

    # returning the slope and offset term
    return (alpha, beta)


# returning y = mx + c values
def f(x, m, c):
    return (m*x + c)
```

Plotting the **scatter plot & the regression line**!

```python
def plot_data(x, y, title_str, img_name):
    # finding the regression line
    (m, c) = linear_regression(x, y)

    plt.scatter(x, y, s=7, c='r', label='scatter plot')
    plt.plot(x, f(x, m, c), label=f'y(cm) = {m.round(8)}x(sec) + {c.round(8)}')
    plt.xlabel('Time (sec)')
    plt.ylabel('Height (cm)')
    plt.grid(alpha=0.5)
    plt.legend()
    plt.title(title_str)
    # plt.savefig(f'{img_name}.png', bbox_inches='tight')
    plt.show()
```
